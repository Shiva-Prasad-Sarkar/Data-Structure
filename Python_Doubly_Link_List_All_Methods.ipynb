```python
#node class
class Node:
    def __init__(self,elem,prev=None,next=None):
        self.prev = prev
        self.elem = elem
        self.next = next
        
```

```python
#from array to dhdcll
def createll(arr):
    head = Node(None)
    temp = head
    head.next = head
    head.prev = head
    for i in range(len(arr)):
        new = Node(arr[i])
        new.prev = head
        head.next = new
        head = new
    new.next = temp
    temp.prev = new
    return temp 
    
#by reverse loop
def createlinklist(arr):
    head = Node(arr[len(arr)-1])
    temp = head
    head.next = head
    head.prev = head
    for i in range(len(arr)-2,-1,-1):
        new = Node(arr[i])
        head.prev = new
        new.next = head
        head = new
    dnode = Node(None)
    dnode.next = head
    head.prev = dnode
    dnode.prev = temp
    temp.next = dnode
    return dnode
    
```

```python
#printing link list
#actual order
def printdhdcll(head):
    print(f'[{id(head.prev)}| {head.elem} |{id(head.next)}]',end='<->')
    temp = head.next
    while temp!=head:
        if temp.next!=head:
            print(f'[{id(temp.prev)}| {temp.elem} |{id(temp.next)}]',end = '<->')
        else:
            print(f'[{id(temp.prev)}| {temp.elem} |{id(temp.next)}]')    
        temp = temp.next
        
#reverse order
def printinreverseorder(head):
    print(f'[{id(head.prev)}| {head.elem} |{id(head.next)}]',end='<->')
    temp = head.prev
    while temp!=head:
        if temp.prev!=head:
            print(f'[{id(temp.prev)}| {temp.elem} |{id(temp.next)}]',end = '<->')
        else:
            print(f'[{id(temp.prev)}| {temp.elem} |{id(temp.next)}]')    
        temp = temp.prev
        
```

```python
#copy
def copyll(head):
    newhead = Node(head.elem)
    var = newhead
    temp = head.next
    while temp!=head:
        newn = Node(temp.elem)    
        newhead.next = newn
        newn.prev = newhead
        newhead = newn
        temp = temp.next
    newn.next = var
    var.prev = newn
    
    return var
```

```python
#reverse
#inplace
def reversein(head):
    temp = head.next
    head.prev,head.next = head.next,head.prev
    while temp!=head:
        temp.next,temp.prev=temp.prev,temp.next
        temp = temp.prev
    return temp,next
    #for getting without dummy head reverse return {temp}

#outplace
def reverseout(head):
    newhead = Node(head.elem)
    var = newhead
    temp = head.next
    while temp!=head:
        newn = Node(temp.elem)    
        newhead.prev = newn
        newn.next = newhead
        newhead = newn
        temp = temp.next
    var.next = newn
    newn.prev = var
    return newn.prev
    #for getting without dummy head reverse return {newn}
    
```

```python
#count elements without dummy head
def counter(head):
    c = 0
    temp = head.next
    while temp!=head:
        c+=1
        temp = temp.next
    return c

def countelem(head,val):
    c = 0
    temp = head.next
    while temp!=head:
        if temp.elem==val:
            c+=1
        temp = temp.next
    return c
    
```

```python
#getnode
def getnode(head,ind):
    c = 0
    temp = head.next
    while temp!=head:
        if c==ind:
            return temp
        c+=1
        temp = temp.next
        
#getelement
def getelem(head,ind):
    c = 0
    temp = head.next
    while temp!=head:
        if c==ind:
            return temp.elem
        c+=1
        temp = temp.next

#getindex
def getindex(head,ele):
    c = 0
    temp = head.next
    while temp!=head:
        if temp.elem == ele:
            return c
        c+=1
        temp = temp.next
```

```python
#insert
def insertcll(head,index,val):
    
    if index>counter(head):
        print('Invalid index')
        return head
    else:
        var = getnode(head,index).prev
        varr = var.next
        new = Node(val,var,varr)
        varr.prev = new
        var.next = new
        return head     
        
#remove value of specific index
def delcll(head,index):
    
    if index>=counter(head):
        print('Invalid index')
        return head
    else:
        temp = head
        var = getnode(head,index).prev
        varr = var.next.next
        var.next = varr
        varr.prev = var
        return temp  
        
#remove element(s)            
def removecll(head,val):
    while head.elem==val:
        if head.elem==val:
            head = delcll(head,0)       
    newhead = head
    temp = newhead.next
    while temp!=newhead:
        if temp.elem==val:
            ind = getindex(newhead,temp.elem)
            var = getnode(head,ind).prev
            varr = var.next.next
            var.next = varr
            varr.prev = var
        temp = temp.next
    return newhead
        
```
